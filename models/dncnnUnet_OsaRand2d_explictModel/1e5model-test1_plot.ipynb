```python
import scipy.io as spio
import numpy as np
import sys, os

import matplotlib.pyplot as plt
import math
```

```python
def plotNNFilter(units, title=None):
    dim = len(units.shape)
    if dim ==3:
        plt.figure(1, figsize=(20,20))
        if title: plt.suptitle(title, fontsize=16)

            
        filters = units.shape[0]
        ncols = int(6)
        nrows = int(math.ceil(filters / ncols) + 1)
        
        for i in range(filters):
            plt.subplot(nrows, ncols, i+1)
            plt.imshow(units[i,:,:], clim=(-3,7), interpolation='none')

        plt.subplots_adjust(left=0.12, bottom=0.12, right=0.9, top=0.88, wspace=0.2, hspace=0.2)
        #plt.subplot_tool()
        plt.show()
        
        
    if dim == 2:
        plt.figure(1, figsize=(3,3))
        if title: plt.title(title, fontsize=16)
        plt.imshow(units, clim=(-3,7), interpolation='none')
        plt.show()
```
