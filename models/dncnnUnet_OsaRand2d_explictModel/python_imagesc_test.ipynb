```python
%matplotlib inline

import scipy.io as spio
import numpy as np
import sys, os
import matplotlib.pyplot as plt
```

```python
cur_mat = spio.loadmat('../dncnn_unet_OsaRand2d/test_results/1e5model-test1.mat', squeeze_me=True)

cur_mat = cur_mat['output_clean']

print cur_mat.shape

maxV = 25.0
cur_mat = cur_mat * maxV;
x = np.exp(cur_mat) - 1.0;

print "\nmin \t max"
print np.amin(x), np.amax(x)

neg_idx = x < 0.0
x[neg_idx] = 1e-8

print "\nmin \t max"
print np.amin(x), np.amax(x)

plt.imshow(x, clim=(-3, 7))
# plt.colorbar()
```
